```python
import numpy as np 
import pandas as pd
from functools import partial
```

Submit form [Baseline from contest](https://www.kaggle.com/code/sharthz23/baseline-from-contest)

```python
df = pd.read_csv('/kaggle/input/baseline-from-contest/submit_batch-512_epochs-25.csv')
df.info(memory_usage='deep')
```

```python
def round_preds(pred_str, prec=4):
    preds = [round(float(s), prec) for s in pred_str.split(',')]
    return ','.join([str(p) for p in preds])
```

```python
s = df['prediction'].iloc[0]
print('Original:')
print(s)
print('Round:')
print(round_preds(s))
```

```python
%%time
df['prediction'] = df['prediction'].map(round_preds)
```

```python
df.info(memory_usage='deep')
```

```python
df.to_csv('round_submit.csv', index=False)
```
